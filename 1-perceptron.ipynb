# Perceptron

```python
import numpy as np
```

## AND Operator

and-operator.png

```python
# The expected inputs to pass into an AND gate.
inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
inputs.shape
```

```python
# The expected outputs. The output will be 1 only when both the inputs are 1.
outputs = np.array([0, 0, 0, 1])
outputs.shape
```

```python
# Neural network is about looking for the right sets of weights.
# We need to start somewhere, in this case, weights are initialised as 0.0.
weights = np.array([0.0, 0.0])
weights
```

```python
# This how we control the speed of learning.
# Too big of a value, it will overshoot.
# Too small of a value, it will take a very long time.
learning_rate = 0.1
```

### Step Function

```python
# A neuron needs 2 things, the dot function and the activation function
def step_function(sum):
    if (sum >= 1):
        return 1
    return 0
```

### Calculate output

```python
def calculate_output(instance):
    s = instance.dot(weights)
    return step_function(s)
```

```python
calculate_output(np.array([[0,0]]))
```

### Train

```python
def train():
    total_error = 1
    # We want it to keep on repeating until the loss is low.
    while (total_error != 0):
        total_error = 0
        for i in range(len(outputs)):
            prediction = calculate_output(inputs[i])
            error = abs(outputs[i] - prediction)
            total_error += error
            if error > 0:
                for j in range(len(weights)):
                    print('Weights before: ' + str(weights[j]))
                    # This updates the weights.
                    # The learning rate is part of the formula to control the speed of learning.
                    weights[j] = weights[j] + (learning_rate * inputs[i][j] * error)
                    print('Weights after: ' + str(weights[j]))
                    print('-----')
#                     print('Weight updated: ' + str(weights[j]))
        print('Total error: ' + str(total_error))
        print('=====')
```

```python
train()
```
